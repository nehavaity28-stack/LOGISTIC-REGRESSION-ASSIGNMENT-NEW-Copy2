# Encoding all the categorical data using One-Hot Encoding.

```python
# Dropping the columns which are not needed for modeling
df_train = df_train.drop(['Name', 'Ticket', 'Cabin'], axis=1)

# Encode Sex (binary)
df_train['Sex'] = df_train['Sex'].map({'male': 0, 'female': 1})

# Handle missing Embarked values
df_train['Embarked'].fillna(df_train['Embarked'].mode()[0], inplace=True)

# One-hot encode Embarked
df_train = pd.get_dummies(df_train, columns=['Embarked'], drop_first=True)

print(df_train.head())
```

```python
from sklearn.model_selection import train_test_split

X = df_train.drop('Survived', axis=1)
Y = df_train['Survived']

X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Build a logistic regression model using appropriate libraries (e.g., scikit-learn).

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(X,Y)
df_train["Y_pred"] = model.predict(X)
df_train["Y_pred"]
```

```python
import numpy as np
from sklearn.metrics import log_loss,accuracy_score

loss = log_loss(Y, df_train["Y_pred"])
print("log loss:", np.round(loss,2))

score = accuracy_score(Y,df_train["Y_pred"])
print("Accuracy_score:", np.round(score,2))
```

```python
import pandas as pd
import pickle
import ipywidgets as widgets
from IPython.display import display

# Load the model
with open('logistic_model.pkl', 'rb') as file:
    model = pickle.load(file)

# Create widgets for all the features the model expects
passenger_id = widgets.IntText(value=1, description='PassengerId:')
pclass = widgets.Dropdown(
    options=[(1, 1), (2, 2), (3, 3)],
    value=1,
    description='Pclass:'
)
sex = widgets.Dropdown(
    options=[('Male', 0), ('Female', 1)],
    value=0,
    description='Sex:'
)
age_slider = widgets.FloatSlider(value=30.0, min=0.0, max=100.0, description='Age:')
sibsp = widgets.IntSlider(value=0, min=0, max=10, description='SibSp:')
parch = widgets.IntSlider(value=0, min=0, max=10, description='Parch:')
fare_slider = widgets.FloatSlider(value=50.0, min=0.0, max=500.0, description='Fare:')
embarked_dropdown = widgets.Dropdown(
    options=[('Southampton', 1), ('Cherbourg', 2), ('Queenstown', 3)],
    value=1,
    description='Embarked:'
)

predict_button = widgets.Button(description='Predict')
output = widgets.Output()

# Display widgets
display(passenger_id, pclass, sex, age_slider, sibsp, parch, fare_slider, embarked_dropdown, predict_button, output)

# Define button click behavior
def on_button_clicked(b):
    with output:
        output.clear_output()
        
# Create a dataframe with the correct feature names
        input_data = pd.DataFrame({
            'PassengerId': [passenger_id.value],
            'Pclass': [pclass.value],
            'Sex': [sex.value],
            'Age': [age_slider.value],
            'SibSp': [sibsp.value],
            'Parch': [parch.value],
            'Fare': [fare_slider.value],
            'Embarked_1': [0],
            'Embarked_2': [0],
            'Embarked_3': [0],})
        
# Set the appropriate Embarked column to 1 based on selection
        embarked_col = f'Embarked_{embarked_dropdown.value}'
        input_data[embarked_col] = 1
        
        # Handle Y_pred if it's required by the model
        # This is unusual - Y_pred is typically an output, not an input
        # You might need to adjust this based on how your model was trained
        if 'Y_pred' in model.feature_names_in_:
            input_data['Y_pred'] = 0  # Default value, adjust as needed
        
        # Make prediction
        try:
            # Ensure columns are in the same order as during training
            input_data = input_data[model.feature_names_in_]
            prediction = model.predict(input_data)
            print(f'Prediction: {prediction[0]}')
        except Exception as e:
            print(f"Error making prediction: {e}")

predict_button.on_click(on_button_clicked)
```
